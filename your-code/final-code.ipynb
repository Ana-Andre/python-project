```python
# define rooms and items
couch = {
    "name": "couch",
    "type": "furniture",
}

queen_bed = {
    "name": "queen bed",
    "type": "furniture",
}

double_bed = {
    "name": "double bed",
    "type": "furniture",
}

dresser = {
    "name": "dresser",
    "type": "furniture",
}

dining_table = {
    "name": "dining table",
    "type": "furniture",
}

door_a = {
    "name": "door a",
    "type": "door",
}

key_a = {
    "name": "key for door a",
    "type": "key",
    "target": door_a,
}

piano = {
    "name": "piano",
    "type": "furniture",
}

game_room = {
    "name": "game room",
    "type": "room",
}

bedroom_1 = {
    "name": "bedroom 1",
    "type": "room",
}

bedroom_2 = {
    "name": "bedroom 2",
    "type": "room",
}

living_room = {
    "name": "living room",
    "type": "room",
}

outside = {
  "name": "outside",
  "type": "room",
}

door_b = {
    "name": "door b",
    "type": "door",
}

door_c = {
    "name": "door c",
    "type": "door",
}

door_d = {
    "name": "door d",
    "type": "door",
}

key_b = {
    "name": "key for door b",
    "type": "key",
    "target": door_b,
}
    
key_c = {
    "name": "key for door c",
    "type": "key",
    "target": door_c,
}
    
key_d = {
    "name": "key for door d",
    "type": "key",
    "target": door_d,
}

all_rooms = [game_room, bedroom_1, bedroom_2, living_room, outside]

all_doors = [door_a, door_b, door_c, door_d]

all_keys = [key_a, key_b, key_c, key_d]


# define which items/rooms are related
object_relations = {
    "game room": [couch, piano, door_a],
    "piano": [key_a],
    "outside": [door_d],
    "door a": [game_room, bedroom_1],
    "bedroom 2": [double_bed, dresser, door_b],
    "double bed": [key_c],
    "dresser": [key_d],
    "bedroom 1": [queen_bed, door_a, door_b, door_c],
    "queen bed": [key_b],
    "door b": [bedroom_1 , bedroom_2],
    "door c": [bedroom_1, living_room],
    "door d": [living_room, outside],
    "living room": [dining_table, door_d, door_c]
}

# define game state. Do not directly change this dict. 
# Instead, when a new game starts, make a copy of this
# dict and use the copy to store gameplay state. This 
# way you can replay the game multiple times.
#later called game_state

INIT_GAME_STATE = {
    "current_room": game_room,
    "keys_collected": [],
    "target_room": outside,
}
```

```python
def linebreak():
    """
    Print a line break
    """
    print("\n\n")
    

#trigger function
#calls play_room(room)
def start_game():
    """
    Start the game
    """
    print("\nYou wake up on a couch and find yourself in a strange house with no windows which you have never been to before. You don't remember why you are here and what had happened before. You feel some unknown danger is approaching and you must get out of the house, NOW!")
    play_room(game_state["current_room"])
    

#############################    
####"""First Challenge"""####
#############################


#main function
#calls explore_room(room)
#calls examine_room(room)
def play_room(room):
    """
    Play a room. First check if the room being played is the target room.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this room) or examine an item found here.
    """
    game_state["current_room"] = room
    #explore = 0
    if(game_state["current_room"] == game_state["target_room"]):
        print("\n")
        print("There's a monster at the door! The monster is obsessed with rock-paper-scissors. In order to get out you need to win the game.")
        game_monster_end(room)
        
        print("Congrats! You escaped the room!")
    
    elif(game_state["current_room"] == game_room):
        inicial_monster(room)        
        
    else:
        print("\nYou are now in " + room["name"])
        explore = 0
        linebreak()
        
        intended_action = input("What would you like to do? Type [1] to Explore or [2] to Examine").lower().strip()
        
        
        if intended_action == "1":
            explore = 1
            explore_room(room)
            examine_item(input("\nWhat would you like to examine?").lower().strip())
        
        elif intended_action == "2" and explore == 1:
            examine_item(input("\nWhat would you like to examine?").lower().strip())
        
        elif intended_action == "2" and explore == 0:
            print("\nYou need to explore the room first.")
            play_room(room)
        
        else:
            print(f"Not sure what you mean by {intended_action}. Type [1] to Explore or [2] to Examine")
            play_room(room)
        linebreak()
        
def inicial_monster(room):
    
    """Creates a option to panic or explore the room
    only if you are in the game room"""
    
    game_state["current_room"] = room
    
    print("\nWhat would you like to do?\n Maybe there is a way to get out for this room\n")
    opt = input("Press [1] to Panic, [2] to Explore the room or [3] to Examine.")
    
    if opt == "1":
        print ("\nYou chose Panic. The monster eats you. \n GAME OVER")
        panic_monster()

    elif opt == "2":
        explore_room(room)
        play_room(room) 
        
    elif opt == "3":
        examine_item(input("\nWhat would you like to examine?").strip())

    else:
        print(f"Sorry, I don't understand {opt}. Can you please try again?")
        inicial_monster()
    
    linebreak()

### What to do if you panic!            

def panic_monster():
    
    """After you panic and lose, the player can chose to end the game or restart it"""
    
    restart = str(input("\nDo you wish to try again? [y/n]"))
    restart = restart.lower()

    if restart == "y":
        start_game()
        
    elif restart == "n":
        print ("\nThe monster is sad and eats you anyway")
        
    else:
        print(f"Sorry, I don't understand {restart}. Can you please try again?")
        panic_monster()
    
    linebreak()
    



def explore_room(room):
    """
    Explore a room. List all items belonging to this room.
    """
    items = [i["name"] for i in object_relations[room["name"]]]
    print("\nYou explore the room. This is " + room["name"] + ". \nYou find " + ", ".join(items) + ".")
    linebreak()
    
    

def get_next_room_of_door(door, current_room):
    """
    From object_relations, find the two rooms connected to the given door.
    Return the room that is not the current_room.
    """
    connected_rooms = object_relations[door["name"]]
    for room in connected_rooms:
        if(not current_room == room):
            return room
    
    linebreak()


def examine_item(item_name):
    """
    Examine an item which can be a door or furniture.
    First make sure the intended item belongs to the current room.
    Then check if the item is a door. Tell player if key hasn't been 
    collected yet. Otherwise ask player if they want to go to the next
    room. If the item is not a door, then check if it contains keys.
    Collect the key if found and update the game state. At the end,
    play either the current or the next room depending on the game state
    to keep playing.
    """
    current_room = game_state["current_room"]
    next_room = ""
    output = None
    
    for item in object_relations[current_room["name"]]:
        if(item["name"] == item_name):
            output = "\nYou examine " + item_name + ". "
            if(item["type"] == "door"):
                have_key = False
                for key in game_state["keys_collected"]:
                    if(key["target"] == item):
                        have_key = True
                if(have_key):
                    output += "\nYou unlock it with a key you have."
                    next_room = get_next_room_of_door(item, current_room)
                else:
                    output += "\nIt is locked but you don't have the key."
            else:
                if(item["name"] in object_relations and len(object_relations[item["name"]])>0):
                    item_found = object_relations[item["name"]].pop()
                    game_state["keys_collected"].append(item_found)
                    output += "\nYou find " + item_found["name"] + "."
                else:
                    print("\nThere isn't anything interesting about it.")
                    examine_item(input("\nWhat would you like to examine?").lower().strip())
                                      
            print(output)
            #explore = 1
            break

    if(output is None):
        print("\nThe item you requested is not found in the current room.")
        examine_item(input("\nWhat would you like to examine?").lower().strip())
    
    if(next_room and input("\nDo you want to go to the next room? [y/n]").lower().strip() == 'y'):
        play_room(next_room)
    else:
        play_room(current_room)
        
    linebreak()
        
        
def game_monster_end(room):
    """
    Once player reaches the living room, he finds a monster! This monster loves playing
    rock-paper-scissors and in order to the player continue his/her quest to the outside
    he needs to win one game with the monster.
    """ 
    game_state["current_room"] = room
    
    import random
    options = ['stone', 'paper', 'scissors']
    number_games = 0
    machine = 0
    human = 0
    max_games = 1
    win = (max_games // 2) + 1
    
    while machine < win or human < win:    

        human_weapon = input("Choose your weapon (stone, paper or scissors): ")
        while (human_weapon != 'stone') and (human_weapon != 'paper') and (human_weapon != 'scissors'):
            print("Weapon not available." + "\n" + "\n" + "Choose from stone, paper or scissors: ")
            human_weapon = input()

        machine_weapon = random.choice(options)
        print("Monster's weapon: ",machine_weapon)

        if (human_weapon == 'stone' and machine_weapon == 'stone') or (human_weapon == 'paper' and machine_weapon == 'paper') or (human_weapon == 'scissors' and machine_weapon == 'scissors'):
            print('Tie')
            print("\n")
        
        elif (human_weapon == 'stone' and machine_weapon == 'scissors') or (human_weapon == 'paper' and machine_weapon == 'stone') or (human_weapon == 'scissors' and machine_weapon == 'paper'):
            print('You won the game! You can get out!')
            human += 1
            print("\n")
            break
        
        elif (machine_weapon == 'stone' and human_weapon == 'scissors') or (machine_weapon == 'paper' and human_weapon == 'stone') or (machine_weapon == 'scissors' and human_weapon == 'paper'):
            print('You lost')
            machine += 1
            print("\n")
        
        if (machine == 0 and human == 0) or (machine ==1):
            continue 
        
        number_games += 1

    linebreak()
```
